```python
# Titok Music search / most used / type Pop
import requests
import time
import pandas as pd

authors = []
albums = []

for i in range(0,10) :

    url = "https://tiktok-all-in-one.p.rapidapi.com/search/music"

    querystring = {"query":"pop","sort":"1","offset":"20"}

    headers = {
        "X-RapidAPI-Key": "aa7a931296mshb6cd8381f6fd6d7p198bd6jsnf57962df27ae",
        "X-RapidAPI-Host": "tiktok-all-in-one.p.rapidapi.com"
    }

    response = requests.request("GET", url, headers=headers, params=querystring)
    results = response.json()
    
    authors.append(results['music'][i]['author'])
    time.sleep(2) #seconds
    albums.append(results['music'][i]['album'])
    time.sleep(2) #seconds

df = pd.DataFrame({
    "author" : authors,
    "album" : albums
})
df
```
